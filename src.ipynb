# Query data

```python
result = collection.find({"Price": {"$lt": 11}})
pd.DataFrame(result)
# for document in result:
#     print(document)
```

```python
def search(attribute,value,condition=None):
  if condition == None:
    search_result = collection.find({attribute:value})
  else:
    search_result = collection.find({attribute:{condition:value}})

  return pd.DataFrame(search_result)
```

## Search by date

```python
search('Date','1951')       #For the year 1951
```

```python
search('Date','1960','$lt')       #For years prior to 1960
```

## Search by Category

```python
search('Category',re.compile(r'fict',re.IGNORECASE))      #Shows all the books whose category contains 'ict' in the string
```

## Search by Price

```python
search('Price',11,'$lt')        #Books that go for less than $11
```

```python
def check_out(collection):
  cart = []
  total = 0

  search_by = input("Enter attribute: ")
  while True:
    value = input(f"Enter {search_by}: ")
    if value == 'done':
      break

    book = collection.find_one({search_by:value})
    if book['Quantity'] == 0:
      print('This book is out of stock')
    else:
      cart.append(book)
      total +=book['Price']


  confirmation = input(f"Total = {total}\nProceed to checkout: (y/n)")
  if confirmation == 'y':
    for book in cart:
      quantity = book["Quantity"]
      if quantity > 0:
        collection.update_one({f"{search_by}": book[f"{search_by}"]},{"$set": {"Quantity": quantity - 1}})
    print("Checkout complete!!")
  else:
    print("checkout aborted!!")
```
